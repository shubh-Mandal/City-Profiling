```python
import matplotlib.pyplot as plt

# Sample production time data for two machines (replace with your actual data)
machine1_data = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
machine2_data = [30, 32, 34, 36, 38, 40, 42, 44, 46, 48]

# Create bar plots
plt.bar(range(len(machine1_data)), machine1_data, label='Machine 1')
plt.bar(range(len(machine2_data)), machine2_data, label='Machine 2')

# Add labels and title
plt.xlabel('Production Run')
plt.ylabel('Production Time')
plt.title('Comparison of Production Time for Two Machines')

# Add legend
plt.legend()

# Show the bar plot
plt.show()
```

```python
# A retail store wants to analyse the sales the sales performance of its products across multiple categories. Generate a 
# barplot to campared the total sales for each product category. Highlight the top performance categories and provide 
# insight to enhance the store product offerings.
```

```python
import matplotlib.pyplot as plt

categories = ['Electronics', 'Clothing', 'Home & Garden', 'Books', 'Toys']
total_sales = [150000, 120000, 180000, 90000, 135000]

plt.figure(figsize=(10, 6))
plt.bar(categories, total_sales, color='skyblue')
        
top_categories = sorted(zip(categories, total_sales), key=lambda x: x[1], reverse=True)[:3]
top_categories_names = [category[0] for category in top_categories]
top_categories_sales = [category[1] for category in top_categories]
plt.bar(top_categories_names, top_categories_sales, color='orange')

# Adding labels and title
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product Category')

# Adding text annotation for top-performing categories
for i in range(len(top_categories)):
    plt.text(i, top_categories_sales[i], str(top_categories_sales[i]), ha='center', va='bottom')
```

```python
# A car dealership is intrested in understanding the relationship between the mileage and resale values of the used cars in
# the inventory. Collect data on the mileage and resale value of the various cars models and create a scatterplot. Analyse
# the plot to identify any correlation pattern and provide insights into how mileage influences resale value, helping the 
# dealership make informed pricing decisions.
```

```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
mileage = [5000, 10000, 15000, 20000, 25000]
resale_value = [25000, 22000, 20000, 18000, 16000]

# Create scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(mileage, resale_value, color='blue', alpha=0.5)

# Add labels and title
plt.xlabel('Mileage (miles)')
plt.ylabel('Resale Value ($)')
plt.title('Relationship Between Mileage and Resale Value')

# Show plot
plt.grid(True)
```
