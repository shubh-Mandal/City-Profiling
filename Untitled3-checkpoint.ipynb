```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
data = pd.read_csv(r"C:\Users\KIIT\Downloads\bhubaneswar.csv")

# Extracting columns for scatter plot
x = data['Total Population (in thousands)']  
y = data['No. of Identfied Slums']  

# Create scatter plot
plt.figure(figsize=(14, 6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Total Population (in thousands)')
plt.ylabel('No. of Identfied Slums')
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the CSV file
df = pd.read_csv(r"C:\Users\KIIT\Downloads\bhubaneswar.csv")

# Sample data (replace with your actual data)
data = {
    'Total Population (in thousands)': df['Total Population (in thousands)'],
    'Population - Male (in thousands)': df['Population - Male (in thousands)'],
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Set the width of the bars
bar_width = 0.35

# Set the position of the bars on the x-axis
r = np.arange(len(df))

# Plotting the bars for each group
plt.figure(figsize=(20, 7))
plt.bar(r, df['Total Population (in thousands)'], color='orange', width=bar_width, edgecolor='grey', label='Total Population')
plt.bar(r + bar_width, df['Population - Male (in thousands)'], color='grey', width=bar_width, edgecolor='grey', label='Population - Male')

# Add xticks on the middle of the group bars
plt.xlabel('Index', fontsize='18')
plt.xticks([r + bar_width/2 for r in range(len(df))], df.index)

# Add y label
plt.ylabel('Values (in thousands)', fontsize='18')

# Add title
plt.title('Clustered Bar Graph', fontsize='18')

# Add legend
plt.legend()

# Show plot
plt.show()
```
